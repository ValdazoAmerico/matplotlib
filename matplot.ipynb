```python
import pandas as pd
fbk = ['Facebook', 2449, True, 2006]
twt = ['Twitter', 339, False, 2006]
ig = ['Instagram', 100, True, 2010]
yt = ['YouTube', 2000, False, 2005]
lkn = ['Linkedin', 663, False, 2003]
wsp = ['Whatsapp', 1600, True, 2009]
lista_rrss = [fbk,twt, ig, yt, lkn, wsp]
df_rrss=pd.DataFrame(lista_rrss, columns=['Nombre','Cantidad','ES_FBK','Año'])
df_rrss
```

```python
from pylab import rcParams
import matplotlib.pyplot as plt
rcParams['figure.figsize'] = 30, 8
plt.rcParams.update({'font.size': 16})
df_rrss_sort=df_rrss.sort_values(by='Cantidad', ascending=False)
lista_col_rrss = ['#3b5998', '#FF0000', '#25d366', '#8a3ab9', '#0e76a8', 'violet']
plt.pie(df_rrss_sort['Cantidad'], labels=df_rrss_sort['Nombre'], colors=lista_col_rrss, autopct='%.2f%%')
plt.show()
```

```python
atletas = pd.read_csv('categorias de corredores.csv', index_col=0)
atletas.info()
```

```python
atletas.head()
```

```python
atletas['Velocidad'].value_counts().plot(kind='pie', autopct='%.2f%%', title='Cantidades de Corredores por Categoria de Velocidad');
```

```python
vel_genero = atletas.groupby(['Velocidad', 'gender']).size()
vel_genero
```

```python
vel_genero = vel_genero.reset_index()
vel_genero
```

```python
vel_genero = pd.pivot_table(vel_genero, columns='Velocidad', index='gender', values=0)
vel_genero
```

```python
vel_genero.plot(kind="bar", title='Velocidad', stacked=True);
```
